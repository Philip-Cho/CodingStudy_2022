# ë¶„í•  ì •ë³µ
- í•˜í–¥ì‹ ì ‘ê·¼ë²•ìœ¼ë¡œ(top-down), ë¬¸ì œë¥¼ ë‚˜ëˆŒ ìˆ˜ ì—†ì„ ë•Œê¹Œì§€ ë‚˜ëˆ„ì–´ì„œ ê°ê°ì„ í’€ë©´ì„œ ë‹¤ì‹œ í•©ë³‘í•˜ì—¬ ë¬¸ì œì˜ ë‹µì„ ì–»ëŠ” ì•Œê³ ë¦¬ì¦˜
- ì¼ë°˜ì ìœ¼ë¡œ ì¬ê·€í•¨ìˆ˜ë¡œ êµ¬í˜„
- Memorization ê¸°ë²• ì‚¬ìš© ì•ˆí•¨
- ë¬¸ì œë¥¼ ìª¼ê°¤ ë•Œ, ë¶€ë¶„ ë¬¸ì œëŠ” ì„œë¡œ ì¤‘ë³µë˜ì§€ ì•ŠìŒ(ë™ì  ê²Œíšë²•ê³¼ì˜ ì°¨ì´)
    - ì˜ˆì‹œ: **ë³‘í•©ì •ë ¬, í€µì •ë ¬**

```python
# í”¼ë³´ë‚˜ì¹˜ ìˆ˜ì—´ (ì¬ê·€)
def fibo(num):
    if num<=1:
        return num
    return fibo(num-1) + fibo(num-2)
```

```python
fibo(4)
```

```python
# í”¼ë³´ë‚˜ì¹˜ ìˆ˜ì—´ (ë™ì ê³„íšë²•) - nì´ í´ ë•Œ ì¬ê·€ë³´ë‹¤ ë” ë¹ ë¦„
# ğŸ”¥num=0ì¼ë•Œ ì—ëŸ¬ ë°œìƒ
def fibo_dp(num):
    cache = [0 for index in range(num+1)]
    cache[0] = 0
    cache[1] = 1
    
    for index in range(2, num+1):
        cache[index] = cache[index - 1] + cache[index-2]
    return cache[num]
```

```python
fibo_dp(3)
```

## [ë¬¸í•­1] 2*n íƒ€ì¼ë§(ë°±ì¤€ 11726)

[ë°±ì¤€ 11726](https://www.acmicpc.net/problem/11726)

```python
'''
[ë‚´ í’€ì´]
ê³µëµë²•:
- íŒ¨í„´ì´ ìˆëŠ” ë¬¸ì œëŠ” 6~8ê°œì •ë„ê¹Œì§€ ë‚˜ì—´í•´ë³´ë©° ì í™”ì‹ ì°¾ì•„ë³´ê¸°
- a_1 = 1, a_2 = 2
- a_n = a_(n-1) + a_(n-2) (ë‹¨,  n>=3)
'''
def quiz1(n):
    # ì°¸ê³ : n>=1ì— ëŒ€í•´ì„œë§Œ ì˜¤ë¥˜ ì—†ìŒ
    cache = [0 for _ in range(n+1)] # ğŸ”¥[ì£¼ì˜] ë°°ì—´ index ì¡°ì‹¬
    cache[0] = 1
    cache[1] = 2
    
    for i in range(2, n+1):
        cache[i] = cache[i-1] + cache[i-2]

    return cache[n-1]%10007
```

```python
quiz1(100)
```

```python
print(quiz1(1))
print(quiz1(2))
print(quiz1(3))
print(quiz1(6))
print(quiz1(9))
```

## [ë¬¸í•­2] íŒŒë„ë°˜ ìˆ˜ì—´(ë°±ì¤€ 9461)

[ë°±ì¤€ 9461](https://www.acmicpc.net/problem/9461)

```python
'''
[ë‚´ í’€ì´]
ê³µëµë²•:
- íŒŒë€ìƒ‰ ì‚¼ê°í˜•ì€ í°ìƒ‰ ì‚¼ê°í˜• ë‘ê°œì˜ í•©, í°ìƒ‰ ì‚¼ê°í˜•ì€ íŒŒë€ìƒ‰ ì‚¼ê°í˜• ë‘ê°œì˜ í•©
- a_1 = 1, a_2 = 1, a_3 = 1, a_4 = 2, a_5 = 2
- a_n = a_(n-1) + a_(n-5) (ë‹¨,  n>=6)
'''
def quiz2(n):
    # ì°¸ê³ : n>=1ì— ëŒ€í•´ì„œë§Œ ì˜¤ë¥˜ ì—†ìŒ
    cache = [0]*101 # ğŸ”¥[ì£¼ì˜] ë°°ì—´ index ì¡°ì‹¬
    cache[0] = 1
    cache[1] = 1
    cache[2] = 1
    cache[3] = 2
    cache[4] = 2
    
    for i in range(5, n+1):
        cache[i] = cache[i-1] + cache[i-5]

    return cache[n-1]
```

```python
print(quiz2(12))
print(quiz2(14))
```
